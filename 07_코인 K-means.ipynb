### 데이터 전처리

```python
df_list=[DOGE, VET, ETC, NEO, BCH, GAS, QTUM, IOST, BSV, ONT, LINK, MANA, SNT, IQ, TRX, LTC, EOS, ETH, ONG, IOTA, ZIL, BTT, XRP, THETA, CRO, XTZ, ADA, ATOM, XEM, BTC, BAT, REP, ELF, AERGO, OMG, STX, EMC, XLM, TFUEL, ENJ, WAVES, BTG, ICX, SXP, TT, LSK, QKC, STEEM, MFT, CVC, KNC, HBAR, SBD, POWR, ARK, LAMB, MOC, CRE, DMT, ZRX, STPT, GLM, STORJ, KMD, BORA, MARO, ADX, STMX, TSHP, ORBS, HUM, MBL, LOOM, WAXP, MVL, GRS, ARDR, SOLVE, LBC, UPP, IGNIS, EDR, STRAX, META, FCT, PXL, MTL, POLY, PUNDIX]
df_list_name=['DOGE', 'VET', 'ETC', 'NEO', 'BCH', 'GAS', 'QTUM', 'IOST', 'BSV', 'ONT', 'LINK', 'MANA', 'SNT', 'IQ', 'TRX', 'LTC', 'EOS', 'ETH', 'ONG', 'IOTA', 'ZIL', 'BTT', 'XRP', 'THETA', 'CRO', 'XTZ', 'ADA', 'ATOM', 'XEM', 'BTC', 'BAT', 'REP', 'ELF', 'AERGO', 'OMG', 'STX', 'EMC', 'XLM', 'TFUEL', 'ENJ', 'WAVES', 'BTG', 'ICX', 'SXP', 'TT', 'LSK', 'QKC', 'STEEM', 'MFT', 'CVC', 'KNC', 'HBAR', 'SBD', 'POWR', 'ARK', 'LAMB', 'MOC', 'CRE', 'DMT', 'ZRX', 'STPT', 'GLM', 'STORJ', 'KMD', 'BORA', 'MARO', 'ADX', 'STMX', 'TSHP', 'ORBS', 'HUM', 'MBL', 'LOOM', 'WAXP', 'MVL', 'GRS', 'ARDR', 'SOLVE', 'LBC', 'UPP', 'IGNIS', 'EDR', 'STRAX', 'META', 'FCT', 'PXL', 'MTL', 'POLY', 'PUNDIX']
```

```python
for i in df_list:
    try:
        i.rename(columns = {'오픈' : '시가'}, inplace = True)
        i.rename(columns = {'시가*' : '시가'}, inplace = True)
        i.rename(columns = {'종가**' : '종가'}, inplace = True)
    except:
        pass
```

### movement 변수 생성

```python
df=pd.DataFrame()
for i in range(len(df_list)):
    try:
        df_list[i]['종가']=df_list[i]['종가'].str.replace(',','').astype('float64')
        df_list[i]['시가']=df_list[i]['시가'].str.replace(',','').astype('float64')
    except:
        pass
    try:
        df_list[i]['종가']=df_list[i]['종가'].str.replace('?','').astype('float64')
        df_list[i]['시가']=df_list[i]['시가'].str.replace('?','').astype('float64')
    except:
        pass

    stock_close=df_list[i]['종가']
    stock_open=df_list[i]['시가']
    movement=stock_close-stock_open
    movement.reset_index(drop=True,inplace=True)
    df[df_list_name[i]]=movement

df=df.iloc[:474]
```

```python
#배열로 변환
movement=np.array(df).T
movement
```

### K-means

```python
# import machine learning libraries
from sklearn.pipeline import make_pipeline
from sklearn.cluster import KMeans

# define normalizer
normalizer = Normalizer()

# create a K-means model with 10 clusters
kmeans = KMeans(n_clusters=3, max_iter=1000)

# make a pipeline chaining normalizer and kmeans
pipeline = make_pipeline(normalizer,kmeans)
```

```python
# fit pipeline to daily stock movements
pipeline.fit(movement)
```

```python
print(kmeans.inertia_)

# Intertia is a score of how close each cluster is, 
# so a lower inertia score is better. 
# In this case, we get a score of 7.71.
```

```python
# predict cluster labels
labels = pipeline.predict(movement)

# create a DataFrame aligning labels & companies
df = pd.DataFrame({'labels': labels, 'companies': df_list_name})

# display df sorted by cluster labels
print(df.sort_values('labels'))
```

### PCA

```python
# PCA
from sklearn.decomposition import PCA 

# visualize the results
reduced_data = PCA(n_components = 2).fit_transform(new)

# run kmeans on reduced data
kmeans = KMeans(n_clusters=3)
kmeans.fit(reduced_data)
labels = kmeans.predict(reduced_data)

# create DataFrame aligning labels & companies
df = pd.DataFrame({'labels': labels, 'companies': df_list_name})

# Display df sorted by cluster labels
print(df.sort_values('labels'))
```

### Elbow Visualize (PCA 후)

### K-means 시각화 (PCA 후)

```python
# Define step size of mesh
h = 0.01

# plot the decision boundary
x_min, x_max = reduced_data[:, 0].min() - 1, reduced_data[:,0].max() + 1
y_min, y_max = reduced_data[:, 1].min() - 1, reduced_data[:,1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain abels for each point in the mesh using our trained model
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)

# define colorplot
cmap = plt.cm.Paired

# plot figure
plt.clf()
plt.figure(figsize=(10,10))
plt.imshow(Z, interpolation='nearest',
 extent = (xx.min(), xx.max(), yy.min(), yy.max()),
 cmap = cmap,
 aspect = 'auto', origin='lower')
plt.plot(reduced_data[:, 0], reduced_data[:, 1], 'k.', markersize=5)


# plot the centroid of each cluster as a white X
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
 marker='x', s=169, linewidth=3,
 color='w', zorder=10)

plt.title('K-Means Clustering on Coin Market Movements (PCA-Reduced Data)')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.show()
```
